## El momento angular de una partícula:
El momento angular de una partícula en mecánica clásica viene dado por el producto vectorial del vertor de posición por el vector momento, es decir $\large \vec{L}=\vec r x \vec p$, de aquí construímos el operador mecánico-cuántico $$\large \hat{L}=\begin{bmatrix} \vec i &\vec j &\vec k \\ \hat x &\hat y &\hat z \\ \hat{p}_x &\hat{p}_y &\hat{p}_z\end{bmatrix}=\vec{i} (\hat{y}\hat{p}_z − \hat{z}\hat{p}_y)+ \vec{j}(\hat{z}\hat{p}_x − \hat{x}\hat{p}_z) + \vec{k}(\hat{x}\hat{p}_y − \hat{y}\hat{p}_x)$$

donde hemos tenido precaución de mantener siempre el orden relativo de las componentes deposición antes de las componentes de momento. La expresión clásica de las componentes se puede convertir así de inmediato en el correspondiente operador cuántico:

$$\large \hat{l}_x= \hat{y}\hat{p}_z - \hat{z}\hat{p}_y = \frac{\hbar}{i}[y\frac{\partial}{\partial z} + z\frac{\partial}{\partial y}]$$
$$\large \hat{l}_y= \hat{z}\hat{p}_x - \hat{x}\hat{p}_z=\frac{\hbar}{i}[z\frac{\partial}{\partial x} + x\frac{\partial}{\partial z}]$$
$$\large \hat{l}_z= \hat{x}\hat{p}_y − \hat{y}\hat{p}_x=\frac{\hbar}{i}[x\frac{\partial}{\partial y} + y\frac{\partial}{\partial x}]$$


y $\hat{l}^2 = \hat{l} \cdot \hat{l}=\hat{l}^{2}_x + \hat{l}^{2}_y + \hat{l}^{2}_z$
La definición de las componentes cartesianas es simétrica frente a la
permutación cíclica x → y → z → x → . . . .

Los operadores que representan componentes cartesianas del momento angular no conmutan entre
sí.

$$\large [\hat{l}_x, \hat{l}_y] = i\hbar l_z, [\hat{l}_y, \hat{l}_z] = i\hbar l_x, [\hat{l}_z, \hat{l}_x] = i\hbar l_y$$.

En cambio, $\hat{l}^2$ conmuta con todas y cada una de las componentes cartesianas:
$$\large [\hat{l}^2, \hat{l}_x] = 0, [\hat{l}^2, \hat{l}_y] = 0, [\hat{l}^2, \hat{l}_z] = 0$$

En conclusion, $\hat{l}^2$ y $\hat{l}_z$ forman un conjunto de operadores compatibles. Las restantes componentes del momento angular no lo son, de modo que en mecánica cuántica no es posible conocer el vector momento angular de una partícula.

El tratamiento del momento angular se facilita empleando coordenadas
esféricas polares. Las derivadas parciales se convierten con ayuda de
la regla de la cadena
\begin{equation}\label{10} \left(\frac{\partial }{\partial \xi}\right)_{y,z}=\left(\frac{\partial }{\partial r}\right)_{\theta,\varphi}\cdot\left(\frac{\partial r}{\partial \xi}\right)_{y,z} +\left(\frac{\partial }{\partial \theta}\right)_{r,\varphi}\cdot\left(\frac{\partial\theta}{\partial \xi}\right)_{y,z}+\left(\frac{\partial }{\partial\varphi}\right)_{r,\theta}\cdot\left(\frac{\partial\varphi}{\partial \xi}\right)_{y,z} \end{equation}    
    
donde $\xi=x, y, z$
    
Ahora sólo falta calcular las derivadas parciales de $r$,$θ$,$φ$ respecto de $x$,$y$,$z$ empleando las ecuaciones de la figura 1.
\begin{equation}\label{18} \left(\frac{\partial r}{\partial x}\right)=sen\theta cos\varphi, \left(\frac{\partial r}{\partial y}\right)=sen\theta sen\varphi, \left(\frac{\partial r}{\partial z}\right)=cos\theta \end{equation}
\begin{equation}\label{23} \left(\frac{\partial\theta}{\partial x}\right)=\frac{cos\theta\cdot cos\varphi}{r}, \left(\frac{\partial\theta}{\partial y}\right) = \frac{cos\theta\cdot sen\varphi}{r}, \left(\frac{\partial\theta}{\partial z}\right) = \frac{-sen\theta}{r} \end{equation}
\begin{equation}\label{27} \left(\frac{\partial\varphi}{\partial x}\right)=-\frac{sen\varphi}{rsen\theta}, \left(\frac{\partial\varphi}{\partial y}\right) = \frac{cos\varphi}{rsen\theta}, \left(\frac{\partial\varphi}{\partial z}\right) = 0 \end{equation}   

Sustituyendo estas nueve derivadas en las ecuaciones de la regla de la cadena, se obtienen los operadores


\begin{eqnarray}  \frac{\partial}{\partial x} &=& sen\theta cos\varphi\cdot\frac{\partial}{\partial r}+\frac{cos\theta cos\varphi}{r}\cdot\frac{\partial}{\partial\theta}-\frac{sen\varphi}{rsen\theta}\cdot\frac{\partial}{\partial\varphi}\\ \frac{\partial}{\partial y} &=& sen\theta sen\varphi\cdot\frac{\partial}{\partial r}+\frac{cos\theta sen\varphi}{r}\cdot\frac{\partial}{\partial\theta}+\frac{cos\varphi}{rsen\theta}\cdot\frac{\partial}{\partial\varphi} \\ \frac{\partial}{\partial z} &=& cos\theta\cdot\frac{\partial}{\partial r}-\frac{sen\theta}{r}\cdot\frac{\partial}{\partial\theta}\label{32} \end{eqnarray}
    
Para finalizar sustituimos las expresiones calculadas en las ecuaciones de $\hat{l}_x$, $\hat{l}_y$ y $\hat{l}_z$, completándose así el cambio a coordenadas esféricas.

\begin{equation} \hat{l}_x =-i\hbar\left[y\frac{\partial}{\partial z}-z\frac{\partial}{\partial y}\right]= \end{equation}
\begin{equation} -i\hbar\left[rsen\theta sen\varphi\left(\not{cos\theta\frac{\partial}{\partial r}} -\frac{sen\theta}{r}\frac{\partial}{\partial\theta}\right)-rcos\theta\left(\not{sen\theta sen\varphi\frac{\partial}{\partial r}} +\frac{cos\theta sen\varphi}{r}\frac{\partial}{\partial\theta}+\frac{cos\varphi}{rsen\theta}\frac{\partial}{\partial\varphi}\right)\right] \end{equation}

Empleando $sen^2θ+cos^2θ=1$ y $cotgθ=\frac{cosθ}{senθ}$ y simplificando:
\begin{equation} \hat{l}_x =i\hbar\left[sen\varphi\frac{\partial}{\partial \theta}+cotg\theta cos\varphi \frac{\partial}{\partial\varphi}\right] \end{equation}
De forma análoga obtenemos el operador $\hat{l}_y$ en coordenadas polares esféricas

\begin{equation} \hat{l}_y =-i\hbar\left[z\frac{\partial}{\partial x}-x\frac{\partial}{\partial z}\right]= \end{equation}
\begin{equation} -i\hbar\left[rcos\theta \left(\not{sen\theta cos\varphi\frac{\partial}{\partial r}} +\frac{cos\theta cos\varphi}{r}\frac{\partial}{\partial\theta}+\frac{sen\varphi}{rsen\theta}\frac{\partial}{\partial\varphi}\right)-rsen\theta cos\varphi\left(\not{cos\theta\frac{\partial}{\partial r}} -\frac{sen\theta}{r}\frac{\partial}{\partial\theta}\right)\right] \end{equation}  

Empleando $sen^2θ+cos^2θ=1$ y $cotgθ=\frac{cosθ}{senθ}$ y simplificando:
\begin{equation} \hat{l}_y =-i\hbar\left[cos\varphi\frac{\partial}{\partial \theta}-cotg\theta sen\varphi \frac{\partial}{\partial\varphi}\right] \end{equation}

Repitiendo el mismo procedimiento obtenemos $\hat{l}_z$ en esféricas.
\begin{equation} \hat{l}_z = -i\hbar\left[{x}\frac{\partial}{\partial y}-{y}\frac{\partial}{\partial x}\right]=-i\hbar\frac{\partial}{\partial\varphi} \end{equation}

También podemos calcular el operador $\hat{l}^2=\hat{l}_x^2+\hat{l}_y^2+\hat{l}_z^2$ en coordendas esféricas
\begin{equation} \hat{l}^2 = -\hbar^2 \left(\frac{\partial^2}{\partial\theta^2}+cotg\theta\frac{\partial}{\partial\theta}+\frac{1}{sen^2\theta}\frac{\partial^2}{\partial\varphi^2}\right) \end{equation}
Obsérvese que los operadores del momento angular son sólo función de las coordenadas $θ$ y $φ$ y no de $r$. Además, como consecuencia de nuestra elección de coordenadas esféricas polares, $\hat{l}_z$ es el momento conjugado de $ϕ$.
Para completar la transformación de coordenadas necesitamos conocer los rangos de las variables y
el elemento de volumen diferencial:

$$dV = r^2\,{\sin}\,\theta\,dr\,d\theta\,d\varphi$$

## 2. El Rotor rígido en Mecánica Cuántica
<p style="text-align:justify">
Si ninguna fuerza actúa sobre rotor, podemos considerar $V = 0$ y el Hamiltoniano se puede escribir así:.
$$\large \hat{H} = \hat{T} + \hat{V} = \frac{\hat{L^2}}{2 I}$$
$$\hat H =- \frac{\hbar^2}{2I} \left [ {1 \over \sin \theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \over \partial \theta} \right) + {1 \over {\sin^2 \theta}} {\partial^2 \over \partial \varphi^2} \right]$$
La ecuación de Schrödinger puede ser escrita así:
$$\hat H \Psi = E \Psi ⇒- \frac{\hbar^2}{2I} \left [ {1 \over \sin \theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Psi \over \partial \theta} \right) + {1 \over {\sin^2 \theta}} {\partial^2 \Psi \over \partial \varphi^2} \right]=E \Psi $$
$$⇒  {1 \over \sin \theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Psi \over \partial \theta} \right) + {1 \over {\sin^2 \theta}} {\partial^2 \Psi \over \partial \varphi^2} +\frac{2IE}{\hbar^2} \Psi = 0 $$

Esta ecuación tiene dos variables angulares y por tanto es posible resolverse por el método de separación de variables:
$$\large Y(\theta , \varphi)=\Theta(\theta)\Phi(\varphi)$$

Sustituyendo $\large Y(\theta , \varphi)=\Theta(\theta)\Phi(\varphi)$ en la ecuación de Schrödinger y multiplicando por $\sin^2 \theta$ (**Separación de variables**)
    
$$\large {\sin \theta \over \Theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Theta \over \partial \theta} \right) +\frac{2IE}{\hbar^2}{\sin^2 \theta} = -{1 \over {\Phi}} {\partial^2 \Phi \over \partial \varphi^2}$$

Esta ecuación se satisface igualando ambos miembros a una constante, por ejemplo $m^2$. Podemos separar en dos ecuaciones diferenciales:    
$$\large -{1 \over {\Phi}} {\partial^2 \Phi \over \partial \varphi^2}=m^2 ⇒ {\partial^2 \Phi \over \partial \varphi^2} + m^2 \Phi=0 $$
$$\large {\sin \theta \over \Theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Theta \over \partial \theta} \right) +\frac{2IE}{\hbar^2}{\sin^2 \theta}=m^2$$
$$\large ⇒{\sin \theta \over \Theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Theta \over \partial \theta} \right) +\frac{2IE}{\hbar^2}{\sin^2 \theta}-m^2=0$$
$$\large ⇒{1 \over \sin \theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \Theta \over \partial \theta} \right) +\left (\beta - \frac{m^2}{\sin^2 \theta}\right)\Theta=0$$

donde $\large \beta=\frac{2I}{\hbar²}E$

La primera Ecuación diferencial tiene un salución simple y es:
$$\large \Phi(\varphi)=Ce^{\pm im \varphi}$$
   
La función $\Phi(φ)$ no es admisible como propia ya que al sumar 2π a φ, estamos en el mismo punto del espacio y la función $\Phi(φ)$ no debe cambiar. Para resolver este problema aplicamos condiciones límite obligando a que $\Phi(φ)$ sea periódica de periodo $2π$.

$$\large \begin{equation} \Phi(\varphi + 2\pi)=\Phi(\varphi) \end{equation}$$
$$\large \begin{equation} Ae^{im(\varphi+2\pi)}=Ae^{im\varphi} \end{equation}$$    
$$\large \not {{Ae^{im\varphi}}}e^{im2\pi}=\not{Ae^{im\varphi}}$$    

Simplificando: $e^{im2π}=1$ Empleando la relación de Euler $e^{iα}=cosα+isenα$

$$\large \begin{equation} e^{im2\pi}=1=cos{(m2\pi)}+isen{(m2\pi)} \end{equation}$$
    
Para que esta última ecuación se cumpla, el coseno debe ser $1$ y el seno $0$. Por tanto $m$ debe ser un entero positivo o negativo
    
$$\begin{equation} m=0,\pm 1, \pm 2,.... \end{equation}$$
$$\begin{equation} \Phi(\varphi)=Ae^{im\varphi}\;\; m=0,\pm 1, \pm 2,.... \end{equation}$$    

La constante C se calcula por normalización    

$$\begin{equation} 1=\int_{0}^{2\pi}\Phi^{\ast}(\varphi)\Phi(\varphi)d\varphi =\int_{0}^{2\pi}A^{\ast}e^{-im\varphi}Ae^{im\varphi}d\varphi=|A|^2 \int_{0}^{2\pi}d\varphi=|A|^2 2\pi \end{equation}$$
$$\begin{equation} A=\frac{1}{\sqrt{2\pi}} \end{equation}$$   
$$\begin{equation} \Phi(\varphi)=\frac{1}{\sqrt{2\pi}}e^{im\varphi}\;\;m=0,\pm 1, \pm 2...... \end{equation}$$    
    
</p>

La segunda Ecuación diferencial se puede transformar, haciendo un cambio de variables:

$$\large x=\cos (\theta), \beta =l(l+1)$$
$$\large (1-x^2){\partial^2 \Theta \over \partial \theta^2}-2x {\partial \Theta \over \partial \theta} + \left (l(l+1) - \frac{m^2}{1-x^2}\right)=0$$    

La Ecuación de arriba es la ecuación asociada de Legendre de orden-lésimo. Por tanto, la solución de Θ es    
$$\large \Theta_\ell^m (\theta) = A P_\ell^m (\cos{\theta} )$$    

donde $P_\ell^m$ es la función asociado de Legendre, definida por    
    
$$\large P_\ell^{m}(x) = (-1)^m (1-x^2)^{m/2} \frac{d^m}{dx^m} \left( P_\ell(x) \right)$$

En términos del $cos(\theta)$ ($x=cos(\theta)$):

$$\large P_\ell^{m}(\cos\theta) = (-1)^m (\sin \theta)^m\ \frac{d^m}{d(\cos\theta)^m}\left(P_\ell(\cos\theta)\right)\$$
    
Las funciones de onda angular son llamadas **armónicos esféricos**:    

$$\large Y_\ell^m (\theta, \varphi ) = N e^{i m \varphi } P_\ell^m (\cos{\theta} )$$

$$\large Y_\ell^m( \theta , \varphi ) = \sqrt{{(2\ell+1)\over 4\pi}{(\ell-m)!\over (\ell+m)!}}  \, P_\ell^m ( \cos{\theta} ) \, e^{i m \varphi }$$

### **Expresiones útiles de los Armónicos Esféricos**:
**Ortogonalidad**:
    
$$\large \int_{\theta=0}^\pi\int_{\varphi=0}^{2\pi}Y_\ell^m \, Y_{\ell'}^{m'}{}^* \, d\Omega=\delta_{\ell\ell'}\, \delta_{mm'} $$
**Conjugado de la función**:

**Forma real de la Función de Onda**

<math display="block">\begin{align}
\large Y_{\ell m} &= \large \begin{cases}
\dfrac{i}{\sqrt{2}} \left(Y_\ell^{m} - (-1)^m\, Y_\ell^{-m}\right) & \text{if}\ m < 0\\
Y_\ell^0 & \text{if}\ m=0\\
\dfrac{1}{\sqrt{2}} \left(Y_\ell^{-m} + (-1)^m\, Y_\ell^{m}\right) & \text{if}\ m > 0.

\end{align}

### **Energía del Rotor Rígido**:
El operador hamiltoniano en cordenadas esféricas polares está dado por:
    
$$\large \hat H =- \frac{\hbar^2}{2I} \left [ {1 \over \sin \theta} {\partial \over \partial \theta} \left ( \sin \theta {\partial \over \partial \theta} \right) + {1 \over {\sin^2 \theta}} {\partial^2 \over \partial \varphi^2} \right] $$

La ecuación de Schrödinger nos permite determinar la energía
 
$$\large \hat H Y_\ell^m (\theta, \varphi) = \frac{\hbar^2}{2I} \ell(\ell+1) Y_\ell^m (\theta, \varphi)$$

Por tanto
$$\large E_\ell = {\hbar^2 \over 2I} \ell \left (\ell+1\right)$$
con $\ell = 0, 1, 2...$ y $m=-\ell,-\ell+1,\dots,0, 1, 2 \dots, +\ell$, de tal forma que para valor dado de $\ell$ existen **$2\ell +1$ estados con la misma energía (estados degenerados)**

Generalmente la energía rotacional se expresa en términos de una constante $B$, **constante rotacional**, así:
$$\large  E_\ell  = B\; \ell \left (\ell+1\right)\quad \textrm{with}\quad B \equiv \frac{\hbar^2}{2I}$$

```python
import matplotlib.gridspec as gridspec
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import sph_harm
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

# Malla de adaos de los ángulos polar y azimutal
theta = np.linspace(0, np.pi, 4000)
phi = np.linspace(0, 2*np.pi, 4000)
# Crea una malla de datos 2-D de los ángulos theta y phi.
theta, phi = np.meshgrid(theta, phi)
# Calcula las coordenadas cartesianas en cada punto de los angulos theta y phi en la malla.
xyz = np.array([np.sin(theta) * np.sin(phi), np.sin(theta) * np.cos(phi), np.cos(theta)])

def plot_Y(ax, l, m): # Definimos una función para la graficación de los Armónicos esféricos
    """Gráfica de los armónicos esféricos de grado l y orden m en el sistema de coordenadas XYZ."""

    # Usamos la librería Scipy para invocar las funciones de los armónicos esféricos 
    Y = sph_harm(abs(m), l, phi, theta)

    # Obtenemos la combinación lineal de los  armónicos esféricos para obtener la forma real de los mismos y poderlos graficar.
    if m < 0:
        Y = np.sqrt(2) * (-1)**m * Y.imag
    elif m > 0:
        Y = np.sqrt(2) * (-1)**m * Y.real
    Yx, Yy, Yz = np.abs(Y) * xyz
    
    # Graficamos usando la función "surface" 
    ax.plot_surface(Yx, Yy, Yz, rstride=50, cstride=50, cmap='jet', alpha=0.8, edgecolors = 'k', linewidth = 0.2)

    # Dibujamos los ejes de referenciaDraw a set of x, y, z axes for reference.
    ax_lim = 0.5
    ax.plot([-ax_lim, ax_lim], [0,0], [0,0], c='0.5', lw=1, zorder=10)
    ax.plot([0,0], [-ax_lim, ax_lim], [0,0], c='0.5', lw=1, zorder=10)
    ax.plot([0,0], [0,0], [-ax_lim, ax_lim], c='0.5', lw=1, zorder=10)
    # Límites y título 
    ax.set_title(r'$Y_{{{},{}}}$'.format(l, m))
    ax_lim = 0.5
    ax.set_xlim(-ax_lim, ax_lim)
    ax.set_ylim(-ax_lim, ax_lim)
    ax.set_zlim(-ax_lim, ax_lim)
    ax.axis('off')

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
l = int(input("Introduzca el valor del número cuántico l = "))
m = int(input("Introduzca el valor del número cuántico m = "))
plot_Y(ax, l, m)
plt.savefig('Y{}_{}.png'.format(l, m))
plt.show()
```
