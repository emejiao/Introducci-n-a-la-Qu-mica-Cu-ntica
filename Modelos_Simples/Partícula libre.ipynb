**<font color=blue size=5> Descripción del modelo Partícula Libre en una dimensión (1D)**

**<div class="alert alert-info"><font color=red size=3> Una partícula libre es aquella partícula que no está sujeta a la influencia de un potencial. Como cualquier otra partícula o sistema, su comportamiento y situación puede ser descrito por su función de onda, $\left|\psi(x)\right>$. Para conocer el comportamiento de la partícula, debemos medir sus propiedades (calcular sus propiedades desde el punto de vista mecánico-cuántico). Podemos utilizar un operador MC, como por ejemplo el operado momento lineal $\hat P$.
El operador momento lineal debe obedecer las siguientes condiciones (Ecuación de valor propio):
$$\hat P \left| \psi_k(x) \right> =p\left | \psi_k(x)\right>  \tag{1}$$
donde *p* es el valor propio o "eigenvalor" del momento (valor real).**

**<font color=blue size=3> En la coordena $x$, el operador momento lineal se representa así: $\hat P =-i\hbar \frac{\partial }{\partial x}$, de tal forma que la ecuación de valor propio es una ecuación diferencial que se resuelve fácilmente para obtener $\left| \psi_k(x) \right>$ como $\psi_k(x) = c\ e^{ikx}$, donde $k$ es un número real.**

**<font color=blue size=3> Veamos:**  
**<font color=blue size=3> $$\hat P \psi_k(x) =p \psi_k(x)$$** 

**<font color=blue size=3> $$-i\hbar \frac{\partial {c\ e^{ikx}}}{\partial x} =-i\hbar\ c\ ik\ e^{ikx} $$** 

**<font color=blue size=3> $$\hbar k\ c\ e^{ikx} = \hbar k\ \psi_k(x) \tag{2}$$**
**<font color=blue size=3> con $p=\hbar k$**

```python
#Podemos resolver la Ec. Diferencial usando sympy
x, hbar, p= sp.symbols('x hbar p') # Definimos variables y parámetros
psi = sp.Function('psi', complex=True)(x) # Definimos la función
Ec_m_pl = sp.Eq(psi.diff(x,1)+(p/(sp.I*hbar))*psi,0) # Expresamos la Ec. diferencial
display(Ec_m_pl) # La mostramos par ver si corresponde 
sol=sp.dsolve(Ec_m_pl,psi) # Se resulve con la función de sympy "dsolve"
sol
```

**<font color=blue size=3> Aunque *$\psi_k(x)$ no puede ser normalizada*, la constante $c$ es escogida de forma que la función aparezca así:
<font color=blue size=3> $$\large{\psi_k(x) = \frac{1}{\sqrt{2\pi}} e^{ikx}}$$**

---
###  <font color=green> Nota sobre la normalización usando la función delta de Dirac  

---
<font color=blue> La normalización de acuerdo con el primer postulado de la MC, se define; $$\left<\psi_{k'}(x)|\psi_k(x)\right> = \left<k'|k \right> = \int^{+\infty}_{-\infty} \psi_{k'}^*(x)\psi_k(x) dx = \int^{+\infty}_{-\infty} c^* e^{-ik'x} ce^{ikx} dx = |c|^2 \int^{+\infty}_{-\infty} e^{i(k-k')x} dx = |c|^2 2\pi\ \delta(k-k')$$

<font color=blue> Donde usamos la ***función delta de Dirac*** que se define:
<font color=blue> $$\delta(x) = \frac{1}{2\pi} \int^{+\infty}_{-\infty} e^{ivx} dv$$
<font color=blue> y tiene las siguientes propiedades:
<font color=blue> $$\delta(x-a)=0\ si  x \neq a$$
<font color=blue> $$\int^\infty_{-\infty} f(x)\ \delta(x-a) \ dx =f(a)$$

<font color=blue> La constante de normalización $c$ sería $c = \frac{1}{\sqrt{2\pi}}$ en tal forma que:
<font color=blue> $$\left<\psi_{k'}(x)|\psi_k(x)\right> = \delta(k-k')$$

**<font color=red size=5> Regresando a la *Partícula libre*.**

**<div class="alert alert-danger"><font color=blue size=3>Una vez conocemos la función de onda podemos hacernos las siguientes preguntas:  
*Cuál es el momento de una partícula libre?, Cuál es es la unidad del momento?, Está cuantizado el momento de una *partícula libre*?**

**<font color=blue size=3> En unidades SI, $k$ tiene unidades de $\frac{1}{m}$ y $\hbar$ tiene unidades de $J.s =\frac{kg.m^{2}}{s}$, por tanto de la ecuación 2 podemos ver que $p=\hbar k$ y las unidades son $\frac{J.s}{m} = \frac{Kg.m}{s}$**

**<font color=blue size=3> Para visualizar la función de onda $\psi_k(x)$ usamos la relación de Euler $[e^{i\theta}=\cos(\theta)+i\sin(\theta)]$ y graficamos por separado la parte Real e Imaginaria de la función de onda (aunque la función de onda sea *compleja*, la $\hat P $ es un operador Hermítico, así que los valores propios son medibles, números reales).**

**<font color=blue size=3> Por otro lado, la *La Densidad de Probabilidad, [$\psi^*_k(x) \psi_k(x)$]* es una constante, puesto que:**

**<font color=blue size=3> $$ \psi^*_k(x) \psi_k(x) = \frac{1}{\sqrt{2 \pi}}e^{-ikx}\frac{1}{\sqrt{2 \pi}}e^{ikx} =\frac{1}{2\pi}$$**

**<font color=red sze=3> Con el codigo de la celda de abajo podemos entrar diferentes valores para el momento $k$ y la longitud espacal $x$** 

```python
# Definimos la función de onda Psi and PsiC (Conjugado Complejo)
def psi(x,k): return (1.0/np.sqrt(2.0*np.pi))*(np.cos(k*x)+np.sin(k*x)*1j)
def psiC(x,k): return (1.0/np.sqrt(2.0*np.pi))*(np.cos(k*x)-np.sin(k*x)*1j)

# Usamos input para que el usuario introduzca el valor de las variables
k = float(input("Introduzca el valor de k (en Angstroms-1) = "))
xmax = float(input("Introduzca el maximo valor de x (en Angstroms) = "))

# Generando la gráfica de la función de onda
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(-xmax, xmax, 1000)
fig, ax = plt.subplots()
lim1=1/np.sqrt(2*np.pi) # Valor máximo de la función de onda
ax.axis([-xmax,xmax,-lim1*1.1,lim1*1.1]) # Definimos los límites de la gráfica
ax.plot(x, psi(x,k).real, label="Real", color="blue") # Gráfica de la parte imaginaria de la función de onda
ax.plot(x, psi(x,k).imag, label="Imag.", color="red") # Gráfica de la parte real de la función de onda
ax.hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black") # Aadicionams una linea horizontal a 0
# Definimos las etiquetas, leyendas, etc
ax.legend(loc=2);ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\psi_k(x)$')
str1=r"$k$ = "+str(k)+r" A$^{-1}$"
plt.title('Función de onda para '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Generando la gráfica de la Densidad de Probabilidad
fig, ax = plt.subplots()
ax.axis([-xmax,xmax,0.0,lim1*lim1*1.4]) # Los limites de la gráfica
ax.plot(x, psi(x,k)*psiC(x,k),label="Probability Density", color="green") # Gráfica de la densidad de la probabilidad
# Etiquetas, leyendas, etc
ax.legend(loc=2);
ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\left|\psi_k(x)\right|^2$')
str1=r"$k$ = "+str(k)+r" A$^{-1}$"
plt.title('Densidad de probabilidad para '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Muestra la gráficaen pantalla
plt.show()
```

**<font color=orange size=4> La primera gráfica muestra que la función de onda es completamente deslocalizada en la coordenada x. Sin embargo, si nos fijamos la segunda grafica vemos que la densidad de probabilidad ($|\psi^*_k(x) \psi_k(x) |$) es la misma.**

**<font color=blue size=4>Veamos ahora, Cuál es la probabilidad de encontrar la particula libre en un $\Delta x$?**

**<font color=red size=3>Esta es una pregunta crucial, puesto que las medidas nunca se hacen en todo el espacio al mismo tiempo, sino que en lugar de eso, nuestros instrumentos realizan las medidas en una sección del espacio, para este caso $1D$, $\Delta x$** 
**<font color=red size=3> $$\Psi^*_k(x) \Psi_k(x) \Delta x = \left|\Psi_k(x)\right|^2 \Delta x = \left(\frac{1}{\sqrt{2\pi}} e^{ikx}\right)^*\frac{1}{\sqrt{2\pi}} e^{ikx} \Delta x = \frac{1}{2\pi} \Delta x $$**

**<font color=red size=3> La probabilidad de encontrar la partícula dentro de un $\Delta x$, en *cualquier* lugar en el espacio, es la misma cuando la partícula tiene un momento, $\hbar k$, bien-definido.**

**<font color=blue size=3> Ahora veamos que ocurre si consideramos un rango de valores de momento $\hbar k$. Consideremos la superposición de ondas con un rango de momento dados por $p=p_0 \pm \Delta p$ $\hspace{0.25cm}$ or $\hspace{0.25cm}$ $\hbar k = \hbar(k_0 \pm \Delta k)$**

**<font color=blue size=3> Puesto que el momento varía continuamente, usamos una integración para la superposicion de ondas, todo igualmente ponderado:   
$$\Psi_{\Delta k}(x) =N\int^{k_o+\Delta k}_{k_o-\Delta k}\ e^{ikx}dk$$
$$\Psi_{\Delta k}(x) = \frac{2N\sin(\Delta kx)}{x} e^{ik_ox}$$
donde $N$ es una constante de normalización y es igual a $N=\frac{1}{2\sqrt{\pi\Delta k}}$**

**<font color=red size=3>  Para entender esta nueva función de onda, gráficamos su parte Real e Imaginario de manera separada.**   

```python
# Definimos las funciones
def psi_contour(x,dk): return (np.sin(dk*x)/(np.sqrt(np.pi*dk)*x))
def psi(x,k,dk): return psi_contour(x,dk)*(np.cos(k*x)+np.sin(k*x)*1j)

# Leyendo la entrada de las variables del usuario
#print 'Para entender es nueva función de onda, gráficamos su parte Real y Imaginaria de manera separada'
k = float(input("Introduzca los valores para k_o (en Angstroms-1) = "))
dk = float(input("Introduzca los valores para Delta k (en Angstroms-1) = "))
xmax = float(input("Introduzca el máximo valor para x (en Angstroms) = "))

# Generando las gráficas
lim1 = np.sqrt(dk/np.pi)
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(-xmax, xmax, 900)
fig, axes = plt.subplots(1, 2, figsize=(13,4))
str1=r"$k_o  \pm \Delta k$ = "+str(k)+r" $\pm$ "+str(dk)+r" A$^{-1}$"
# Sub_gráficas a la izquierda
axes[0].axis([-xmax,xmax,-1.1*lim1,1.1*lim1])
axes[0].plot(x, psi(x,k,dk).real, label="Real", color="blue", linewidth=1.8)
axes[0].plot(x, psi_contour(x,dk), label="", linestyle ="--",color="blue", linewidth=1.2)
axes[0].plot(x, -psi_contour(x,dk), label="", linestyle ="--",color="blue", linewidth=1.2)
axes[0].hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black")
axes[0].set_xlabel(r'$x$ (Angstroms)')
axes[0].set_ylabel(r'$Real(\Psi_{\Delta k}(x))$')
axes[0].set_title('Contribución Real a $\Psi_{\Delta k}(x)$ \n for '+str1)
# Sub_gráficas a la derecha
axes[1].axis([-xmax,xmax,-1.1*lim1,1.1*lim1])
axes[1].plot(x, psi(x,k,dk).imag, label="Imag.", color="red", linewidth=1.8)
axes[1].plot(x, psi_contour(x,dk), label="", linestyle ="--",color="red", linewidth=1.2)
axes[1].plot(x, -psi_contour(x,dk), label="", linestyle ="--",color="red", linewidth=1.2)
axes[1].hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black")
axes[1].yaxis.set_label_position("right")
axes[1].set_xlabel(r'$x$ (Angstroms)')
axes[1].set_ylabel(r'$Imag(\Psi_{\Delta k}(x))$')
axes[1].set_title('Contribución Imaginaria a $\Psi_{\Delta k}(x)$ \n for '+str1)

# Generando gráficas de la densidad de probabilidad
fig, ax = plt.subplots()
ax.axis([-xmax,xmax,0.0,lim1*lim1*1.1]) # Definiend los límites de la gráficas
ax.plot(x, (psi(x,k,dk).real)**2+(psi(x,k,dk).imag)**2,label="Densidad de Probabilidad", color="green") 
# Etiquetas, leyendas, etc
ax.legend(loc=2);
ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\left|\Psi_{\Delta k}(x)\right|^2$')
plt.title('Densidad de Probabilidad \n para '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Muestra la gráficas
plt.show()
```

**<font color=blue size=3>  Si $k_o > \Delta k$, los componentes $\cos(k_o x)$ y $\sin(k_o x)$ oscilan con altos valores de frecuencias ($k_o$) que modulados por una componente sinusoidal de baja frecuencia ($\Delta k$).**   

**<font color=blue size=3>  Si $\Delta k$ es muy pequeña, la partícula es todabía deslocalizada. Mientras que cuando $\Delta k$ crece, la partícula aparece más localizada en una región específica alrededor de $x = 0$. Esto nos lleva a concluir que cuando la *incertidumbre* en el momento se incrementa, (grandes $\Delta k$), la *incertidumbre* en la posición decrese, y la partícula parece más *localizada*.**

**<font color=green size=3>  La superposición de ondas puede ser conseguida con diferentes contribuciones a diferentes valores del momento. Por ejemplo, podemos construir la superposición con funciones Gausianas. En esta caso, modulamos la contribución de cada k-onda con un valor dado por una distribución Gausiana (la contribución máxima de $k_o$, con pequeñas contribuciones de ondas con otros valores de $k$).**  

**<font color=green size=3> $$\Psi_{\Delta k}(x) =N\int^{+\infty}_{-\infty}{e^{-\frac{(k-k_o)^2}{2\Delta k^2}} \   e^{ikx}}dk,$$ donde $\Delta k$ está relacionando con la anchura de la distribución Gaussiana.**

**<font color=green size=3> Después de la integración, obtenemos:**

**<font color=green size=3> $$\Psi_{\Delta k}(x) = \sqrt{2\pi}N\Delta k e^{-\frac{1}{2}x^2\Delta k^2} e^{ik_ox}$$**
 
**<font color=green size=3> donde $N$ es una constante de normalización y es igual a $N=\frac{1}{\sqrt{2\Delta k\sqrt{\pi^3}}}$**

```python
# Definimos las funciones
def psi_contour(x,dk): return dk*np.exp(-0.5*x*x*dk*dk)/np.sqrt(dk*np.sqrt(np.pi))
def psi(x,k,dk): return psi_contour(x,dk)*(np.cos(k*x)+np.sin(k*x)*1j)

# Leyendo la entrada de las variables del usuario
print ("Gráficas de los componentes Real e Imaginario")
k = float(input("Introduzca el valor de k_o (en Angstroms-1) = "))
dk = float(input("Introduzca el valor de Delta k (en Angstroms-1) = "))
xmax = float(input("Introduzca el valor de x (en Angstroms) = "))

# Generando las gráficas
lim1 = dk/np.sqrt(dk*np.sqrt(np.pi))
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(-xmax, xmax, 900)
fig, axes = plt.subplots(1, 2, figsize=(13,4))
str1=r"$k_o  \pm \Delta k$ = "+str(k)+r" $\pm$ "+str(dk)+r" A$^{-1}$"
# Sub_gráficas a la izquerda
axes[0].axis([-xmax,xmax,-1.1*lim1,1.1*lim1])
axes[0].plot(x, psi(x,k,dk).real, label="Real", color="blue")
axes[0].plot(x, psi_contour(x,dk), label="", linestyle ="--",color="blue", linewidth=1.8)
axes[0].plot(x, -psi_contour(x,dk), label="", linestyle ="--",color="blue", linewidth=1.8)
axes[0].hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black")
axes[0].set_xlabel(r'$x$ (Angstroms)')
axes[0].set_ylabel(r'$Real(\Psi_{\Delta k}(x))$')
axes[0].set_title('Contribución de la parte Real a $\Psi_{\Delta k}(x)$ \n para '+str1)
# Sub_gráficas a la derecha
axes[1].axis([-xmax,xmax,-1.1*lim1,1.1*lim1])
axes[1].plot(x, psi(x,k,dk).imag, label="Imag.", color="red")
axes[1].plot(x, psi_contour(x,dk), label="", linestyle ="--",color="red", linewidth=1.8)
axes[1].plot(x, -psi_contour(x,dk), label="", linestyle ="--",color="red", linewidth=1.8)
axes[1].hlines(0.0, -xmax, xmax, linewidth=1.8, linestyle='--', color="black")
axes[1].set_xlabel(r'$x$ (Angstroms)')
axes[1].set_ylabel(r'$Imag(\Psi_{\Delta k}(x))$')
axes[1].set_title('Contribución de la parte Imaginaria a $\Psi_{\Delta k}(x)$ \n para '+str1)

# Generando gráficas de la densidad de probabilidad
fig, ax = plt.subplots()
ax.axis([-xmax,xmax,0.0,lim1*lim1*1.1]) # Defining the limits to be plot in the graph
ax.plot(x, (psi(x,k,dk).real)**2+(psi(x,k,dk).imag)**2,label="Probability Density", color="magenta") # Plotting the probability density
# Etiquetas, leyendas, etc
ax.legend(loc=2);
ax.set_xlabel(r'$x$ (Angstroms)')
ax.set_ylabel(r'$\left|\Psi_{\Delta k}(x)\right|^2$')
plt.title('Densidad de Probabilidad \n para '+str1)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.0)

# Se muestran la gráficas
plt.show()
```

**<font color=red size=5>Energía de la Partícula Libre**

**<font color=blue size=3>Como se definió el modelo de la partícula libre, ésta no tiene potencial (no actúa sobre ella ninguna fuerza), así que el Hamiltoniano solamente contiene el término energía cinética. La Ecuación de Schrödinger sería:
$$\large -\frac{\hbar^2}{2m} \frac{\partial^2\psi(x)}{\partial x^2} =E\psi(x)$$ 
Puesto que se conocen las funciones de onda funciones propias del operador momomento, la determinación del valor propio de la energía se puede determinar fácilmente a partir de las propiedades de conmutación.**

**<font color=blue size=3> La propiedad de conmutación de operadores expresa que si dos operadores ($\hat A$ y $\hat B$) conmutan, entonces ellos comparten un conjunto común de funciones propias. 
Veamos si el Hamiltoniano y el momento lineal conmutan: $\left[\hat P, \hat H \right] =0$?.   
$\hat H =\frac{\hat P^2}{2m}$, evaluemos:**

<font color=blue> $$\left[\hat P, \hat H \right] = \hat P \cdot \hat H - \hat H \cdot \hat P$$ 
<font color=blue> $$\hat P \cdot \frac{\hat P^2}{2m} - \frac{\hat P^2}{2m}\cdot \hat P = \frac {1}{2m}\left(\hat P \cdot \hat P^2-\hat P \cdot \hat P^2 \right) = 0$$

**<font color=blue size=3> Los dos operadores conmutan, llevando así a la existencia de un conjunto completo de funciones propias comunes a los dos operadores.
Como se conocen la funciones propias de $\hat P$, podemos tratar de ver que también son funciones propias del operador Hamiltoniano, $\hat H$. para $\left|k\right> = \frac{1}{\sqrt{2\pi}}e^{ikx}$, evaluamos el resultado de $\hat H \left|k\right>$:**

<font color=blue> $$\hat H\left|k\right> = -\frac{\hbar^2}{2m} \frac{\partial^2\left|k\right>}{\partial x^2}
= -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2}\left(\frac{1}{\sqrt{2\pi}}e^{ikx} \right)$$

<font color=blue> $$\hspace 1.5cm = -\frac{\hbar^2}{2m\sqrt{2\pi}} \frac{\partial^2e^{ikx}}{\partial x^2}
= -\frac{\hbar^2}{2m\sqrt{2\pi}} \left(ik\right)^2 e^{ikx}$$ 

<font color=blue> $$= \frac{\hbar^2k^2}{2m} \left(\frac{1}{\sqrt{2\pi}}e^{ikx} \right)  = \frac{p^2}{2m}\left|k\right>$$

**<font color=blue size=3> El primero y el último término en esta derivación muestra que las funciones propias del momento son funciones propias del Hamiltoniano, con valores propios iguales a la forma clásica de la energía cinética. Sin embargo, hay que ser cuidadoso en el análisis de este comportamiento, aunque la energía se ve como el de una partícula *clásica* el comportamiento es muy diferente. Para sistemas cuánticos, las partículas *nunca pueden estar en reposo* (si $\Delta k = 0$ la particula estará completamente deslocalizada; si la partícula está bien localizada, entoces podrá tener diferentes valores de momento).**
