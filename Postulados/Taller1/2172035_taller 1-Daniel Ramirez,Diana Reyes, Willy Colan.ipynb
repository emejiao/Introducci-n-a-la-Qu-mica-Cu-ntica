<div class="alert alert-danger">Demuestre que en el intervalo 0 ≤ θ ≤ 2π, las funciones $\large e^{inθ}$, donde n=0, ±1, ±2, ..., forman un conjunto de funciones ortogonales. Halle la constante de normalización para este conjunto de funciones.

<div class="alert alert-info"><font color='blue'> Si b=0, ±1, ±2,..., pero diferente de n. Definimos las siguientes funciones:
    
$$\large f(x)=e^{in\theta}$$
    
$$\large g(x)=e^{ib\theta}$$
    
Para evaluar si este conjunto de funciones es ortogonales definimos el conjugado de la funcion f(x) e integramos el producto de g(x) y el conjugado de f(x) en el intervalo 0 ≤ θ ≤ 2π y para que sea un conjunto de funciones ortogonales el resultado de la integral debe ser 0.
    
$$\large f^{*}(x)=e^{-in\theta}$$
    
$$\large \int_0^{2\pi} e^{-in\theta}*e^{ib\theta} \mathrm{d}\theta = \int_0^{2\pi} e^{i\theta(b-n)} \mathrm{d}\theta$$
    
Por defincion de n y b tenemos b-n = entero = E.
    
$$\large \int_0^{2\pi} e^{i E \theta} \mathrm{d}\theta$$
    
Aplicando una sustitucion simple en la integral. $$\large iE\theta=u$$
$$\large iEd\theta=du$$ 
$$\large d\theta=\frac{du}{iE}$$

Obtenemos entonces la siguiente integral.
    

$$\large \frac{1}{iE} \int_0^{i2E\pi} e^{u} \mathrm{d}u= \frac{1}{iE}(e^{i2E\pi}-1)$$
    
    
Aplicando reglas de los numeros imaginarios $\large e^{ia\pi}=(-1)^a$ obtenemos.
    
$$\large \frac{1}{iE}((-1)^{2E}-1)=0$$
    
Con este resultado aseguramos que la potencia 2E sea un entero par y por tanto, aseguramos que $\large (-1)^{2E}-1$ sea cero y por tanto el resultado de la integral es cero, sin importar que E sea negativo, por tanto.
    
$$\large \int_0^{2\pi} e^{-in\theta}*e^{ib\theta} \mathrm{d}\theta= 0$$

Con el resultado de la integral se demuestra que el conjunto de funciones $\large e^{inθ}$ son ortogonales en el intervalo 0 ≤ θ ≤ 2π.

<div class="alert alert-info"><font color='blue'> Ahora hallamos la consante de normalizacion de $\large e^{inθ}$ asi.
    
$$\large \int_0^{2\pi} e^{in\theta}*e^{-in\theta} \mathrm{d}\theta=\large \int_0^{2\pi} 1 \mathrm{d}\theta= 2\pi $$
    
Obtenemos la constante de normalizacion asi.
    
$$\large A=\frac{1}{\sqrt{2\pi}}=\frac{\sqrt{2}}{2\sqrt{\pi}}$$
